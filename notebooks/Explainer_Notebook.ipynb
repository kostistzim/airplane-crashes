#### Rename Columns

#### Parse and Correct Dates

The `accident_date` column needs to be converted from text strings into proper datetime objects for time-based analysis. We also implement a correction for ambiguous two-digit years (like '68' potentially being parsed as 2068 instead of 1968, this error occured in the first iteration of our code).

1.  Parse dates using the `DD-Mon-YY` format (`%d-%b-%y`). Any dates that cannot be parsed are converted to `NaT` (Not a Time).
2.  Remove rows with unparseable dates (`NaT`), as the date is crucial.
3.  Adjust years that were parsed as being in the future (Year > Current Year + 1). For these, we assume they belong to the previous century and subtract 100 years.

```python
# Parse the 'accident_date' column from text strings to datetime objects and correct ambiguous years.

# Step 2a: Initial parsing using the 'DD-Mon-YY' format. Errors are turned into NaT (Not a Time).
df_corrected['accident_date'] = pd.to_datetime(df_corrected['accident_date'], format='%d-%b-%y', errors='coerce')

# Step 2b: Drop rows where the date could not be parsed (became NaT).
initial_rows = len(df_corrected)
df_corrected.dropna(subset=['accident_date'], inplace=True)
print(f"Dropped {initial_rows - len(df_corrected)} rows with invalid date format.")

# Step 2c: Adjust years parsed as being in the future.
current_year = pd.Timestamp.now().year + 1 # Buffer for current year
future_mask = df_corrected['accident_date'].dt.year > current_year
df_corrected.loc[future_mask, 'accident_date'] = df_corrected.loc[future_mask, 'accident_date'] - pd.DateOffset(years=100)
print(f"Adjusted years for {future_mask.sum()} records initially parsed as future dates.")
print("Date parsing and correction complete.")
print(f"Corrected Date range: {df_corrected['accident_date'].min()} to {df_corrected['accident_date'].max()}")
```

#### Extract Year and Month

To facilitate grouping and time-based analysis (like finding trends per year), we extract the year and month from the corrected `accident_date` into separate numerical columns.

```python
# Extract the Year and Month from the corrected 'accident_date'.
df_corrected['accident_year'] = df_corrected['accident_date'].dt.year
df_corrected['accident_month'] = df_corrected['accident_date'].dt.month
print("Extracted Year and Month from accident_date.")
```

#### Clean Fatalities Column

The original `fatalities` column was text ('object' type) and contained non-numeric values (e.g., '?') and missing data. To use it for calculations, we perform the following:

1.  Convert the column to a numeric type. Any values that cannot be converted (errors) become `NaN` (Not a Number).
2.  Calculate the median number of fatalities (excluding `NaN`s).
3.  Fill the missing values (`NaN`s) with the median. We use the median because it's less affected by extremely high-fatality accidents compared to the mean. The median value of 0 suggests that a majority of recorded incidents had no fatalities.
4.  Convert the cleaned column to an integer type.

```python
# Clean the 'fatalities' column.

# Step 3a: Convert to numeric, coercing errors (non-numeric values) into NaN.
df_corrected['fatalities'] = pd.to_numeric(df_corrected['fatalities'], errors='coerce')

# Step 3b: Calculate the median number of fatalities (excluding NaNs).
median_fatalities = df_corrected['fatalities'].median()

# Step 3c: Fill missing values (NaNs) in 'fatalities', not 'damage'
df_corrected['fatalities'] = df_corrected['fatalities'].fillna(median_fatalities)

# Step 3d: Convert the column to integer type now that NaNs are handled.
df_corrected['fatalities'] = df_corrected['fatalities'].astype(int)

print(f"Fatalities column cleaned (non-numeric converted, NaNs filled with median: {median_fatalities}).")
```

#### Clean Damage Column

We handle the few missing values (`NaN`) in the `damage` column by filling them with the string 'Unknown' to ensure consistency.

```python
# Clean the 'damage' column.

# Fill missing values (NaNs) with the string 'Unknown'.
df_corrected['damage'] = df_corrected['damage'].fillna('Unknown')

print("Damage column cleaned (NaNs filled with 'Unknown').")
```

#### Verify Cleaning

Let's check the number of missing values remaining in each column after the cleaning process. This helps confirm our steps and identify any columns still needing attention.

```python
# Display the count of remaining missing values per column.
print("--- Data Cleaning Finished ---")
print("\nFinal missing values count:")
print(df_corrected.isnull().sum())
```

# interactive plot using bokeh

```python
# Import necessary Bokeh modules
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.models import HoverTool, LinearColorMapper, ColorBar
from bokeh.palettes import Viridis256 # Or another sequential palette
from bokeh.transform import transform
from bokeh.io import output_notebook, output_file
import pandas as pd
import numpy as np

# --- Prepare Data for Decade/Month Heatmap ---
# Assuming df_corrected DataFrame exists and is cleaned as before

print("Preparing data for Decade/Month heatmap...")
# Create a 'decade' column
df_heatmap_decade = df_corrected.copy()
df_heatmap_decade['decade'] = (df_heatmap_decade['accident_year'] // 10) * 10

# Group by decade and month, count accidents
heatmap_agg = df_heatmap_decade.groupby(['decade', 'accident_month']).size().reset_index(name='count')

# Convert decade and month to strings for categorical axes
heatmap_agg['decade_str'] = heatmap_agg['decade'].astype(str) + "s" # e.g., "1980s"
heatmap_agg['month_str'] = heatmap_agg['accident_month'].apply(lambda x: f"{x:02d}") # Format as '01', '02', etc.

print("Data preparation complete.")


# --- Create the Interactive Decade/Month Heatmap ---
print("Generating interactive Bokeh Decade/Month heatmap...")

# Setup output context (choose one or adapt as needed)
# output_notebook()
# output_file("heatmap_decade_month.html")

# Create ColumnDataSource
source = ColumnDataSource(heatmap_agg)

# Define the ranges for axes (sorted list of unique string values)
decades = sorted(heatmap_agg['decade_str'].unique(), reverse=True) # Reverse for recent decades at top
months = sorted(heatmap_agg['month_str'].unique()) # Should be '01' through '12'

# Define color mapper
min_count = heatmap_agg['count'].min()
max_count = heatmap_agg['count'].max()
# Using Viridis256 palette, reversed so high counts are yellow/bright
# Consider LogColorMapper if counts vary extremely across decades:
# from bokeh.models import LogColorMapper
# color_mapper = LogColorMapper(palette=Viridis256[::-1], low=min_count, high=max_count)
color_mapper = LinearColorMapper(palette=Viridis256[::-1], low=min_count, high=max_count)


# Define Hover Tool
hover = HoverTool(
    tooltips=[
        ('Decade', '@decade_str'),
        ('Month', '@month_str'),
        ('Accidents', '@count'),
    ]
)

# Create the figure
p = figure(
    height=350, # Can be shorter now with fewer rows
    width=800,
    x_range=months,
    y_range=decades, # Use the decade strings range
    tools=[hover, "pan", "wheel_zoom", "box_zoom", "reset", "save"],
    title="Number of Accidents per Month and Decade",
    x_axis_label="Month",
    y_axis_label="Decade",
    toolbar_location="below",
    x_axis_location="above"
)

# Add rect glyphs for the heatmap
p.rect(
    x='month_str',
    y='decade_str', # Use decade string for y-axis
    width=1,
    height=1,
    source=source,
    fill_color=transform('count', color_mapper),
    line_color=None
)

# Add Color Bar
color_bar = ColorBar(color_mapper=color_mapper,
                     label_standoff=12,
                     location=(0,0))
p.add_layout(color_bar, 'right')

# Customize appearance
p.axis.axis_line_color = None
p.axis.major_tick_line_color = None
p.xaxis.major_label_orientation = np.pi / 4 # Rotate month labels

# Show the plot
show(p)

print("Bokeh Decade/Month heatmap code generated.")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.models import HoverTool
from bokeh.palettes import Category10
from bokeh.layouts import column
from bokeh.io import output_notebook

# Enable notebook output
output_notebook()

# --- Load and Prepare Data ---
# Note: Assuming df_corrected exists. Replace with actual data loading if needed.
# For testing, you can create a sample df_corrected if not available:
"""
# Sample data for testing (uncomment if df_corrected is not defined)
data = {
    'fatalities': [0, 10, 5, 0, 20],
    'damage': ['Minor', 'Destroyed', 'Major', 'Unknown', 'Major'],
    'operator': ['US Air Force', 'Delta Airlines', 'Private Pilot', 'Lufthansa', 'Navy'],
    'accident_date': ['2020-01-01', '2019-06-15', '2018-03-10', '2021-11-22', '2020-07-19'],
    'accident_year': [2020, 2019, 2018, 2021, 2020]
}
df_corrected = pd.DataFrame(data)
"""
print("Using existing df_corrected DataFrame for Military vs. Civilian analysis.")

# Verify required columns
required_cols = ['fatalities', 'damage', 'operator', 'accident_year']
if not all(col in df_corrected.columns for col in required_cols):
    raise ValueError("df_corrected is missing required columns: " + str([col for col in required_cols if col not in df_corrected.columns]))

# Create a copy with required columns
df_loc = df_corrected[required_cols].copy()

# Ensure correct data types
df_loc['fatalities'] = pd.to_numeric(df_loc['fatalities'], errors='coerce').fillna(0).astype(int)
df_loc['damage'] = df_loc['damage'].fillna('Unknown').astype(str)
df_loc['operator'] = df_loc['operator'].fillna('Unknown').astype(str)
df_loc['accident_year'] = pd.to_numeric(df_loc['accident_year'], errors='coerce').fillna(0).astype(int)

# Drop rows with invalid accident_year
df_loc = df_loc[df_loc['accident_year'] > 0]

# --- 1. Categorize Operators ---
print("Categorizing operators...")

# Define keywords
military_kws = ['Air Force', 'Army', 'Navy', 'Luftwaffe', 'Military', 'Guard', 'Armada', 'Marines',
                'Fuerza Aerea', 'Ejercito', 'Armee de l Air', 'RAF', 'Defense']
private_kws = ['Private', 'Individual']
civilian_kws = ['Airlines', 'Airways', 'Cargo', 'Express', 'Avianca', 'Aeroflot', 'Lufthansa',
                'Air ', 'Avia', 'Lines', 'Servicios', 'Aerovias', 'Lineas', 'Transportes',
                'Transair', 'Pan Am']

# Apply categorization
def categorize_operator(op):
    op = str(op).lower()
    if any(kw.lower() in op for kw in military_kws):
        return 'Military'
    elif any(kw.lower() in op for kw in private_kws):
        return 'Private'
    elif any(kw.lower() in op for kw in civilian_kws):
        return 'Civilian/Commercial'
    return 'Other/Unknown'

df_loc['op_category'] = df_loc['operator'].apply(categorize_operator)
print("Operator categorization complete.")

# DIAGNOSTIC 1: Category counts
print("\n<<< DIAGNOSTIC 1: Category Counts BEFORE Aggregation >>>")
print(df_loc['op_category'].value_counts())

# --- 2. Aggregate Data ---
print("\nAggregating accident counts by year and category...")
df_agg_counts = df_loc.groupby(['accident_year', 'op_category']).size().reset_index(name='count')
print("Aggregation complete.")

# DIAGNOSTIC 2: Unique categories after aggregation
print("\n<<< DIAGNOSTIC 2: Unique Categories AFTER Aggregation >>>")
agg_categories = df_agg_counts['op_category'].unique()
print(agg_categories)

# DIAGNOSTIC 3: Check Military data
print("\n<<< DIAGNOSTIC 3: Aggregated Data for 'Military' Category >>>")
military_agg = df_agg_counts[df_agg_counts['op_category'] == 'Military']
if not military_agg.empty:
    print(military_agg)
else:
    print("No aggregated data for 'Military'. Check categorization or data.")

# --- 3. Visualize with Bokeh ---
print("\nGenerating Bokeh faceted line chart...")

# Get unique categories and assign colors
if len(agg_categories) == 0:
    raise ValueError("No categories found after aggregation. Cannot plot.")

num_categories = len(agg_categories)
colors = Category10[max(3, min(num_categories, 10))]
if num_categories > len(colors):
    colors = [colors[i % len(colors)] for i in range(num_categories)]
category_colors = dict(zip(agg_categories, colors))

# Define HoverTool
hover = HoverTool(tooltips=[
    ("Year", "@accident_year"),
    ("Category", "@op_category"),
    ("Accidents", "@count"),
])

# Create plots
plots = []
p_first = None

print("\n<<< DIAGNOSTIC 4: Categories being plotted >>>")
for category in agg_categories:
    print(f"  - Processing plot for category: {category}")
    category_data = df_agg_counts[df_agg_counts['op_category'] == category]
    if category_data.empty:
        print(f"    - Skipping category '{category}' (no data).")
        continue

    source = ColumnDataSource(category_data)
    
    # Create figure, only link x_range if p_first exists
    p = figure(height=250, width=800, 
               tools=[hover, "pan", "wheel_zoom", "box_zoom", "reset", "save"],
               title=f"Accidents per Year: {category}",
               x_axis_label="Year", 
               y_axis_label="Number of Accidents")
    
    # Link x_range for subsequent plots
    if p_first:
        p.x_range = p_first.x_range

    # Add glyphs
    color = category_colors[category]
    p.line(x='accident_year', y='count', source=source, color=color, line_width=2)
    p.circle(x='accident_year', y='count', source=source, color=color, size=4)

    # Customize
    p.xgrid.grid_line_color = None
    p.yaxis.axis_label_text_font_style = "bold"

    plots.append(p)
    if not p_first:
        p_first = p

# Show layout
if plots:
    show(column(*plots))
    print("\nBokeh faceted plot generated.")
else:
    print("\nNo plots generated. Check data or aggregation.")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import plotly.express as px

# --- Load and Prepare Data ---
# Assuming df_corrected and df_loc (with op_category) exist from previous steps.
print("Using existing df_loc DataFrame with 'op_category'.")

# Ensure required columns exist and have correct types
required_cols = ['accident_year', 'op_category', 'damage']
if not all(col in df_loc.columns for col in required_cols):
    raise ValueError("df_loc is missing required columns ('accident_year', 'op_category', 'damage'). Please re-run previous steps.")
if not pd.api.types.is_numeric_dtype(df_loc['accident_year']):
    df_loc['accident_year'] = pd.to_numeric(df_loc['accident_year'], errors='coerce')
    df_loc = df_loc.dropna(subset=['accident_year'])
    df_loc['accident_year'] = df_loc['accident_year'].astype(int)
df_loc['op_category'] = df_loc['op_category'].astype(str)
df_loc['damage'] = df_loc['damage'].astype(str)


# --- 1. Feature Engineering: Define Eras ---
print("Defining aviation eras...")

def assign_era(year):
    if year < 1960:
        return '1: Propeller (<1960)'
    elif year >= 1960 and year < 1980:
        return '2: Early Jet (1960-1979)'
    else: # year >= 1980
        return '3: Modern (>=1980)'

df_loc['era'] = df_loc['accident_year'].apply(assign_era)
print("Eras assigned. Value counts:")
print(df_loc['era'].value_counts().sort_index().to_markdown(numalign="left", stralign="left"))


# --- 2. Calculate Damage Proportions ---
print("\nCalculating proportion of damage types within each Era and Operator Category...")

# Group by era and op_category, then get the normalized value counts for damage type
damage_proportions = df_loc.groupby(['era', 'op_category'])['damage'].value_counts(normalize=True).reset_index(name='proportion')

print("Proportion calculation complete. Sample data:")
print(damage_proportions.head().to_markdown(index=False, numalign="left", stralign="left"))


# --- 3. Visualize with Plotly Express ---
print("\nGenerating interactive faceted bar chart...")

# Determine order for damage categories for consistent coloring/stacking
# Define a sensible order if possible (e.g., from least to most severe, plus others)
damage_order = ['non', 'min', 'sub', 'w/o', 'mis', 'unk', 'Unknown']
# Filter proportions data to include only categories present, maintaining the order
existing_damage_cats = damage_proportions['damage'].unique()
ordered_cats_in_data = [cat for cat in damage_order if cat in existing_damage_cats]


# Create faceted bar chart using Plotly Express
# X-axis: Operator Category
# Y-axis: Proportion (should sum to 1 for each bar if barmode='stack')
# Color: Damage Type
# Facet: Era
fig = px.bar(damage_proportions,
             x='op_category',
             y='proportion',
             color='damage',
             facet_col='era', # Separate columns for each era
             category_orders={'damage': ordered_cats_in_data, # Ensure consistent color/stack order
                              'era': sorted(damage_proportions['era'].unique())}, # Sort eras chronologically
             labels={'op_category': 'Operator Category',
                     'proportion': 'Proportion of Accidents',
                     'damage': 'Damage Level',
                     'era': 'Aviation Era'},
             title='Distribution of Damage Severity by Operator Category Across Eras',
             height=500 # Adjust height as needed
             # barmode='stack' is the default when color is used
             )

# Improve layout - format y-axis as percentage, adjust titles
fig.update_layout(yaxis_tickformat='.0%') # Format y-axis ticks as percentages
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1])) # Clean facet titles
fig.update_xaxes(title_text='Operator Category') # Ensure x-axis title is present

# Show the plot
fig.show()
```

### Aviation Eras: Propellers to Jets

To understand how aviation safety evolved with technological changes, we grouped crashes into three eras:

- **Propeller Era** (before 1960)  
- **Early Jet Era** (1960–1979)  
- **Modern Jet Era** (1980–present)  

We then analyzed how crash **damage severity** differs by **operator type** (Civilian vs Military) across these eras. The result is shown below using a faceted bar chart.


```python
# 1. Map Numeric Damage Levels to Categories
# Convert 'damage' column to numeric (if not already)
df_corrected['damage'] = pd.to_numeric(df_corrected['damage'], errors='coerce')

# Map numeric values to severity labels
def categorize_damage_level(d):
    if pd.isnull(d):
        return 'Unknown/None'
    elif d == 0.0:
        return 'Unknown/None'
    elif d == 1.0:
        return 'Minor'
    elif d == 2.0:
        return 'Substantial'
    else:
        return 'Write-off'

df_corrected['damage_cleaned'] = df_corrected['damage'].apply(categorize_damage_level)

# 2. Assign Aviation Era Based on Date
# Make sure date is in datetime format
df_corrected['accident_date'] = pd.to_datetime(df_corrected['accident_date'], errors='coerce')

# Assign era
def assign_era(date):
    if pd.isnull(date):
        return 'Unknown'
    year = date.year
    if year < 1960:
        return 'Propeller Era'
    elif year < 1980:
        return 'Early Jet Era'
    else:
        return 'Modern Jet Era'

df_corrected['era'] = df_corrected['accident_date'].apply(assign_era)

# 3. Create Grouped Bar Chart with Plotly
import plotly.express as px

# Filter out rows without era or damage category
df_plot = df_corrected.dropna(subset=['era', 'damage_cleaned'])

# Create the plot
fig = px.histogram(
    df_plot,
    x="damage_cleaned",
    color="era",
    barmode="group",
    category_orders={
        "era": ["Propeller Era", "Early Jet Era", "Modern Jet Era"],
        "damage_cleaned": ["Write-off", "Substantial", "Minor", "Unknown/None"]
    },
    title="Aircraft Accident Severity by Era",
    labels={"damage_cleaned": "Damage Type", "count": "count"},
    color_discrete_map={
        "Propeller Era": "#66c2a5",
        "Early Jet Era": "#fc8d62",
        "Modern Jet Era": "#8da0cb"
    }
)

fig.update_layout(
    height=600,
    width=800,
    plot_bgcolor='rgba(240, 248, 255, 0.8)',  # Light blue background
    xaxis=dict(title_font=dict(size=14)),
    yaxis=dict(title_font=dict(size=14))
)

# Save the chart as an HTML file
fig.write_html("aircraft_accident_severity.html")

# Optional: also display the chart in the notebook if running in a Jupyter environment
fig.show()
```
